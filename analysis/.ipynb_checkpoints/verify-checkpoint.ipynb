```python
from pymongo import MongoClient
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
holdings = ['000985', '000297', '001178', '001562', '001718', '166002', '460005', '519069', '159905', '163415', '001766', '519688', '519736',
           '006308','004394']
```

```python
fund_code = '006308'
with MongoClient('localhost') as client:
    db = client['fund']
    clc = db['bars']
    res = clc.find({'fund_code':fund_code})
    df = pd.DataFrame(res)
    plot_df = df[['datetime', 'net_value_unit', 'net_value_acc']].set_index('datetime').copy()
    plot_df['net_value_unit'] = plot_df['net_value_unit'].apply(float)
    plot_df['net_value_acc'] = plot_df['net_value_acc'].apply(float)
```

```python
print(fund_code)
plot_df.plot(figsize=(20, 10), title=fund_code)
```
